```python
# Exemplo de preparação de dados
def prepare_mmm_data(data, media_cols, control_cols, target_col):
    # Dividir em treino/teste
    train_size = int(0.8 * len(data))
    train_data = data.iloc[:train_size]
    test_data = data.iloc[train_size:]

    # Separar features e target
    X_media_train = train_data[media_cols].values
    X_control_train = train_data[control_cols].values
    y_train = train_data[target_col].values

    X_media_test = test_data[media_cols].values
    X_control_test = test_data[control_cols].values
    y_test = test_data[target_col].values

    # Normalizar dados
    media_scaler = StandardScaler()
    target_scaler = StandardScaler()

    X_media_train_scaled = media_scaler.fit_transform(X_media_train)
    y_train_scaled = target_scaler.fit_transform(y_train.reshape(-1, 1)).flatten()

    X_media_test_scaled = media_scaler.transform(X_media_test)
    y_test_scaled = target_scaler.transform(y_test.reshape(-1, 1)).flatten()

    return {
        'X_media_train': X_media_train,
        'X_media_train_scaled': X_media_train_scaled,
        'X_control_train': X_control_train,
        'y_train': y_train,
        'y_train_scaled': y_train_scaled,
        'X_media_test': X_media_test,
        'X_media_test_scaled': X_media_test_scaled,
        'X_control_test': X_control_test,
        'y_test': y_test,
        'y_test_scaled': y_test_scaled,
        'media_scaler': media_scaler,
        'target_scaler': target_scaler
    }
```
